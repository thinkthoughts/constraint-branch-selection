# 01 Branch Selection Demo

A minimal toy model showing:

- one observable gives many solutions
- adding constraints narrows the parameter space
- multiple branches may still survive
- an extra orthogonal constraint can select a unique branch


## Setup

We scan a 2-parameter space `(x, y)` and impose simple toy constraints.

- **Observable 1**: points near a circle
- **Constraint 2**: points near a diagonal line
- **Constraint 3**: points near a horizontal line

This creates a clean branch-selection example without claiming any direct physical model.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter grid
x = np.linspace(-2.5, 2.5, 600)
y = np.linspace(-2.5, 2.5, 600)
X, Y = np.meshgrid(x, y)

# Tolerances
eps1 = 0.06
eps2 = 0.12
eps3 = 0.10

# Constraint 1: circle of radius sqrt(2)
c1 = np.abs(X**2 + Y**2 - 2.0) < eps1

# Constraint 2: diagonal line y = x
c2 = np.abs(Y - X) < eps2

# Constraint 3: horizontal line y = 1
c3 = np.abs(Y - 1.0) < eps3

# Combined masks
m12 = c1 & c2
m123 = c1 & c2 & c3

print('Constraint 1 points:', int(c1.sum()))
print('Constraint 1 + 2 points:', int(m12.sum()))
print('Constraint 1 + 2 + 3 points:', int(m123.sum()))
```

## Plot 1 — One observable, many solutions

The first observable leaves a large degenerate set of solutions.


```python
plt.figure(figsize=(7, 7))
plt.contourf(X, Y, c1.astype(float), levels=[-0.5, 0.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Observable 1: many valid solutions')
plt.axis('equal')
plt.show()
```

## Plot 2 — Add a second constraint

The second constraint reduces the space dramatically, but two branches remain.


```python
plt.figure(figsize=(7, 7))
plt.contourf(X, Y, m12.astype(float), levels=[-0.5, 0.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Observable 1 + Constraint 2: narrowed, still degenerate')
plt.axis('equal')
plt.show()
```

## Plot 3 — Add an orthogonal constraint

A third independent constraint selects a single surviving branch.


```python
plt.figure(figsize=(7, 7))
plt.contourf(X, Y, m123.astype(float), levels=[-0.5, 0.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Observable 1 + 2 + 3: unique branch selected')
plt.axis('equal')
plt.show()
```

## Summary

- **Fit** does not imply **identifiability**
- **Constraint stacking** narrows parameter space
- Multiple branches can survive even after good fits
- Extra orthogonal information can select a unique solution

This notebook is a toy demonstration of branch selection in inverse problems.
