# 03 Observability Threshold (v2)

When does degeneracy break?

**Goal:** show how increasing constraint precision collapses multiple branches into one.

> identifiability ∝ constraint strength × precision

```python
import numpy as np
import matplotlib.pyplot as plt

# grid
x = np.linspace(-2.5, 2.5, 600)
y = np.linspace(-2.5, 2.5, 600)
X, Y = np.meshgrid(x, y)

# base structure
circle = np.abs(X**2 + Y**2 - 2) < 0.05
diag = np.abs(Y - X) < 0.12

def constraint(width):
    return np.abs(X - 1.0) < width
```

## Degeneracy → narrowing → identifiability


```python
plt.figure(figsize=(12,4))

widths = [0.6, 0.25, 0.06]
labels = [
    'degenerate (2 branches)',
    'ambiguous (still 2)',
    'identifiable (single branch)'
]

for i, w in enumerate(widths):
    plt.subplot(1,3,i+1)
    
    mask = circle & diag & constraint(w)
    
    plt.contourf(X, Y, mask.astype(float), levels=[-0.5,0.5,1.5])
    
    # mark branches with correct colors
    if i < 2:
        plt.scatter([-1], [-1], s=70, color='tab:blue', label='wrong')
        plt.scatter([1], [1], s=70, color='tab:orange', label='true')
    else:
        plt.scatter([1], [1], s=70, color='tab:orange', label='true')
    
    plt.title(labels[i])
    plt.axis('equal')
    
    if i == 0:
        plt.legend(loc='upper right', frameon=True)

plt.suptitle('Observability threshold → precision resolves degeneracy')
plt.show()
```

## Interpretation

- weak → multiple viable branches
- medium → narrowed but still degenerate
- strong → identifiable solution

**Key:**
constraint strength determines whether branches remain distinguishable.
