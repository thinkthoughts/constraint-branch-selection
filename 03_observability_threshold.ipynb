# 03 Observability Threshold

Toy model for:

- when do two branches become distinguishable?
- how much signal / precision is needed to resolve degeneracy?

Key idea:

> identifiability depends on constraint strength and precision

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2.5, 2.5, 500)
y = np.linspace(-2.5, 2.5, 500)
X, Y = np.meshgrid(x, y)

# base constraint (circle)
circle = np.abs(X**2 + Y**2 - 2) < 0.05

# two candidate branches (diagonal)
diag = np.abs(Y - X) < 0.1

# true vs wrong separated by vertical constraint strength
def constraint(strength):
    return np.abs(X - 1.0) < strength
```

## Vary constraint strength

As constraint tightens, degeneracy collapses.


```python
strengths = [0.6, 0.3, 0.1]

plt.figure(figsize=(12,4))

for i, s in enumerate(strengths):
    plt.subplot(1,3,i+1)
    mask = circle & diag & constraint(s)
    plt.contourf(X, Y, mask.astype(float), levels=[-0.5,0.5,1.5])
    plt.title(f"constraint width = {s}")
    plt.axis('equal')

plt.suptitle('Observability threshold: narrowing constraint resolves branches')
plt.show()
```

## Interpretation

- weak constraint → multiple branches remain
- medium → narrowed but ambiguous
- strong → single identifiable solution

This models:

> identifiability ∝ constraint strength (precision)
